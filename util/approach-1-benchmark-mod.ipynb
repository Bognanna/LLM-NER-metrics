```python
import os
import json
```

```python
path = '../data/external/cadec'
names = os.listdir(f'{path}/text')

with open('../data/external/benchmark/approach_1_benchmark.json') as f:
    benchmark_dataset = json.load(f)
    benchmark_dataset = benchmark_dataset['Data']

new_benchmark = []

for e in benchmark_dataset:
    
    found_name = []
    entry = {
        "file": None,
        "A": [],
        "B": [],
        "C": []
    }

    for name in names:
        with open(f'{path}/text/{name}') as f:
            if e['Sentence'] in f.read():
                found_name.append(name)
                entry['file'] = f'{name[:-3]}ann'
  
    if found_name:
        # y.append((e['Sentence'], f'{found_name[0][:-3]}ann'))
        with open(f'{path}/original/{found_name[0][:-3]}ann') as f:
            for entity in e['Perfect']:
                category = None
                for line in f:
                    if entity in line:
                        category = line.split()[1]

                entry["A"].append(tuple((entity, category)))
    else:
        for entity in e['Perfect']:
            entry["A"].append((entity, None))

    for entity in e['Good']:
        entry["B"].append((entity, None))
    
    for entity in e['Ok']:
        entry["C"].append((entity, None))

    new_benchmark.append(entry)
```

```python
def save_to_jsonl(data, file_path):
    with open(file_path, 'w+') as f:
        for entry in data:
            f.write(json.dumps(entry) + '\n')
```

```python
benchmark_path = '../data/external/benchmark/approach-1-2.json'

save_to_jsonl(new_benchmark, benchmark_path)
```
